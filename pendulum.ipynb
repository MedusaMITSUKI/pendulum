```julia
# パッケージの読み込み
using PyPlot
using Elliptic
```

```julia
x = 0 : 90 # 振り幅 [deg]
g = 9.8    # 重力加速度 [m/s^2]
l = 1.0    # 弦の長さ [m]
```

```julia
T(x) = 4 * sqrt(l / g) * Elliptic.K(sin((x / 2.0) * pi / 360)) # 振り子の周期
```

```julia
plot(x, T.(x))
xlabel("振り幅 [deg]", fontname="Meiryo")
ylabel("周期 [s]", fontname="Meiryo")
```

```julia
T0(x) = 2 * pi * sqrt(l / g) # 振り子の周期 (近似解)
```

```julia
plot(x, T0.(x))
xlabel("振り幅 [deg]", fontname="Meiryo")
ylabel("周期 [s]", fontname="Meiryo")
```

```julia
h(x) = T.(x) / T0.(x) # 振り子の周期の比
```

```julia
plot(x, h.(x))
xlabel("振り幅 [deg]", fontname="Meiryo")
ylabel("周期の比", fontname="Meiryo")
```
