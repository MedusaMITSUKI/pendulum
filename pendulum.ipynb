```julia
# パッケージの読み込み
using PyPlot
using Elliptic
```

```julia
# 変数・定数定義
x = range(0, 90, length = 100) # 振り幅 [deg]
g = 9.8                        # 重力加速度 [m/s^2]
l = 1.0                        # 弦の長さ [m]
```

```julia
# 振り子の周期 (微小振動)
T0(x) = 2 * pi * sqrt(l / g)
```

```julia
# 振り子の周期
T(x) = 4 * sqrt(l / g) * Elliptic.K(sin((x / 2.0) * (pi / 180)) ^ 2)
```

```julia
# 振り子の周期のグラフ
plot(x, T.(x), linewidth = 2)

xlabel("振り幅 [deg]", fontname = "Meiryo", fontsize = 20)
ylabel("周期 [s]", fontname="Meiryo", fontsize = 20)
xticks(fontsize = 16)
yticks(fontsize = 16)
```

```julia
# 第１種完全楕円積分のグラフ
k = range(0, 1.0, length = 100)
plot(k, Elliptic.K.(k), linewidth = 2)

xlabel("k", fontname = "Meiryo",  fontsize = 20)
ylabel("K(k)", fontname = "Meiryo",  fontsize = 20)
xticks(fontsize = 16)
yticks(fontsize = 16)
```

```julia
# 振り子の周期の比
h(x) = T.(x) / T0.(x)
```

```julia
# 振り子の等時性の破れの確認
plot(x, h.(x), linewidth = 2)

xlabel("振り幅 [deg]", fontname = "Meiryo", fontsize = 20)
ylabel("周期の比", fontname = "Meiryo", fontsize = 20)
xticks(fontsize = 16)
yticks(fontsize = 16)
```

```julia
# 振り幅90 [deg] のときの振り子の等時性の破れ
h(90)
```
